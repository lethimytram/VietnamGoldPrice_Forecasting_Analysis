# Gold Price Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Đọc dữ liệu
df_clean = pd.read_csv("gold_prices_cleaned.csv")
# Vì kiểu dữ liệu cột Date đang ở dạng object
# -> Chuyển kiểu dữ liệu cột Date sang datetime
df_clean['Date'] = pd.to_datetime(df_clean['Date'])
df_clean.dtypes
# Đảm bảo dữ liệu được sắp xếp theo thời gian
df_clean = df_clean.sort_values(by='Date')
```

## Line Plot: Price Trend Over Time

```python
# Thiết lập trục thời gian làm chỉ số
df_clean.set_index('Date', inplace=True)

# Vẽ biểu đồ xu hướng giá mua và bán theo thời gian
plt.figure(figsize=(12, 6))
plt.plot(df_clean.index, df_clean['Buy Price'], label='Giá mua', color='blue')
plt.plot(df_clean.index, df_clean['Sell Price'], label='Giá bán', color='orange')
plt.title('Xu hướng giá vàng theo thời gian')
plt.xlabel('Thời gian')
plt.ylabel('Giá (nghìn đồng/lượng)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def plot_boxplot_gold_prices(file_path):
    df_clean = pd.read_csv(file_path)
    df_clean['Date'] = pd.to_datetime(df_clean['Date'])
    df_clean['Year'] = df_clean['Date'].dt.year

    plt.figure(figsize=(12, 6))
    sns.boxplot(x="Year", y="Buy Price (x1000đ/lượng)", data=df_clean, color="lightblue")
    sns.boxplot(x="Year", y="Sell Price (x1000đ/lượng)", data=df_clean, color="lightcoral")

    plt.xlabel("Năm")
    plt.ylabel("Giá vàng (x1000đ/lượng)")
    plt.title("Phân bố giá vàng theo năm")
    plt.xticks(rotation=45)
    plt.grid(True)

    plt.show()

# Gọi hàm
plot_boxplot_gold_prices("gold_prices.csv")
```

## Histogram: Phân phối dữ liệu

```python
# Phân phối giá mua
sns.histplot(df_clean['Buy Price'], kde=True, color='blue', bins=30)
plt.title('Phân phối giá mua vàng')
plt.xlabel('Giá mua (nghìn đồng/lượng)')
plt.ylabel('Tần suất')
plt.show()

# Phân phối giá bán
sns.histplot(df_clean['Sell Price'], kde=True, color='orange', bins=30)
plt.title('Phân phối giá bán vàng')
plt.xlabel('Giá bán (nghìn đồng/lượng)')
plt.ylabel('Tần suất')
plt.show()
```

## Scatter Plot: Mối quan hệ giữa giá mua và giá bán

```python
# Biểu đồ scatter plot giữa giá mua và giá bán
sns.scatterplot(x=df_clean['Buy Price'], y=df_clean['Sell Price'], color='green')
plt.title('Mối quan hệ giữa Giá mua và Giá bán')
plt.xlabel('Giá mua (nghìn đồng/lượng)')
plt.ylabel('Giá bán (nghìn đồng/lượng)')
plt.grid(True)
plt.show()
```
